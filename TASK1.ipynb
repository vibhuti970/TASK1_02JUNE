```python
df_movies = df[df['type'] == 'Movie'].copy()
df_tvshows = df[df['type'] == 'TV Show'].copy()
```

```python

df_tvshows['seasons'] = (
    df_tvshows['duration']
    .str.replace(r' seasons?', '', regex=True, case=False)  
    .str.strip() 
    .astype(float)
    .astype(int)
)
df_tvshows.head(10)
```

```python
import numpy as np

# 1. Remove 'season' or 'seasons' (case-insensitive)
cleaned = df_tvshows['duration'].str.replace(r' seasons?', '', regex=True, case=False)

# 2. Strip leading/trailing spaces
cleaned = cleaned.str.strip()

# 3. Replace empty strings with NaN (in case any remain)
cleaned = cleaned.replace('', np.nan)

# 4. Convert to numeric with errors coerced to NaN (so invalid parsing won't crash)
cleaned_numeric = pd.to_numeric(cleaned, errors='coerce')

# 5. Fill NaN with 0 or another default
cleaned_filled = cleaned_numeric.fillna(0).astype(int)

# 6. Assign back to dataframe
df_tvshows['seasons'] = cleaned_filled

# 7. Sort by seasons
df_tvshows = df_tvshows.sort_values(by='seasons').reset_index(drop=True)
```

```python
print("Movies sorted by minutes:")
print(df_movies[['title', 'duration']])

print("\nTV Shows sorted by number of seasons:")
print(df_tvshows[['title', 'duration']])
```
